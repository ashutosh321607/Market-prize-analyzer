```python

from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
```

```python
SSD_v=[]
HDD_v=[]
RAM_v=[]
GR_v=[]
Name_v=[]
Price_v=[]
Lap_ID=[]
Link_v=[]

#intial values

url="https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree"
j=0
while(1):
    
    page = requests.get(url=url)
    print(url)
    # soup
    soup = BeautifulSoup(page.text)
    lap_list=soup.find_all('div',{'class':'_3O0U0u'})
    nxt_l=soup.find_all('a',{'class':'_3fVaIS'})
    
    if(j==0):
        t=BeautifulSoup(str(nxt_l[0]),'html.parser')
        url='https://www.flipkart.com'+t.a['href']
        j+=1
    else:
        try:
            t=BeautifulSoup(str(nxt_l[1]),'html.parser')
        except IndexError as e:
            break
        url='https://www.flipkart.com'+t.a['href']
    for i in range(len(lap_list)):
        Lap_ID.append(str(lap_list[i])[35:51])
        Price_v.append((lap_list[i].find('div',{'class':'_1vC4OE _2rQ-NK'}).get_text())[1:])
        temp=lap_list[i].find('div',{'class':'_3wU53n'}).get_text()
        Name_v.append(temp)
        t=temp.find('(')
        t2=temp.find(')')
        if((t2-t)<8):
            
            temp=temp[t2+1: ]
            t=temp.find('(')
        try :
            if(int(temp[t+1])):
                strt=t+1

        except ValueError as l:
            temp=temp[t+1: ]
            t=temp.find('(')
            strt=t+1
            
        end=temp.find('Grap')
        if(end>0):
            temp=temp[strt:end+8]
        else:
            end=temp.find(')')
            temp=temp[strt:end]

        temps = re.findall(r'\d+', temp)
        temps=list(map(int,temps))
        if 10 in temps:
            temps.remove(10)
        
        temps.sort()
        
        #laptop link
        link_t=BeautifulSoup(str(lap_list[i].find('a')),'html.parser')
        Link_v.append(('https://www.flipkart.com'+link_t.a['href']))
    
        s_t=temp.find('SSD')
        h_t=temp.find('HDD')
        g_t=temp.find('Grap')    
        if (s_t>0 and h_t>0 and g_t>0):
            HDD_v.append(temps[0])
            GR_v.append(temps[1])
            RAM_v.append(temps[2])
            SSD_v.append(temps[3])
        elif(s_t>0 and h_t>0 and g_t<0):
            HDD_v.append(temps[0])
            GR_v.append(0)
            RAM_v.append(temps[1])
            SSD_v.append(temps[2])
        elif(s_t<0 and h_t>0 and g_t>0):
            HDD_v.append(temps[0])
            GR_v.append(temps[1])
            RAM_v.append(temps[2])
            SSD_v.append(0)
        elif(s_t<0 and h_t>0 and g_t<0):
            if(temps[1]>50):
                HDD_v.append(temps[1])
                GR_v.append(0)
                RAM_v.append(temps[0])
                SSD_v.append(0)
            else:
                HDD_v.append(temps[0])
                GR_v.append(0)
                RAM_v.append(temps[1])
                SSD_v.append(0)
        elif(s_t>0 and h_t<0 and g_t>0):
            HDD_v.append(0)
            if (1 in temps):
                SSD_v.append(temps[0])
                GR_v.append(temps[1])
                RAM_v.append(temps[2])
            else:
                GR_v.append(temps[0])
                RAM_v.append(temps[1])
                SSD_v.append(temps[2])
        elif(s_t>0 and h_t<0 and g_t<0):
            HDD_v.append(0)
            GR_v.append(0)
            if(1 in temps): 

                SSD_v.append(temps[0])
                RAM_v.append(temps[1])
            else:
                RAM_v.append(temps[0])
                SSD_v.append(temps[1])        
        else:
            
                if(len(temps)== 2):
                    HDD_v.append(temps[1])
                    GR_v.append(0)
                    RAM_v.append(temps[0])
                    SSD_v.append(0)
                elif (len(temps)==1):
                    HDD_v.append(0)
                    GR_v.append(0)
                    RAM_v.append(temps[0])
                    SSD_v.append(0) 
                else:
                    HDD_v.append(temps[1])
                    GR_v.append(0)
                    RAM_v.append(temps[0])
                    SSD_v.append(temps[2])
```

```python
Boff=[]
Price_t=list(df['Price'])
Link_t=list(df['Link'])
for i in range(len(Link_t)):
    url=Link_t[i]
    print(i)
    page = requests.get(url=url)
    soup = BeautifulSoup(page.text)
    offer_t=soup.find_all('li',{'class':'_2-n-Lg col'})
    if(len(offer_t)>0):
        T=[]
        for j in range(len(offer_t)):
            t=offer_t[j].get_text()
            p=t.find('Bank Offer')
            pt=t.find('%')
            if(p>=0 and pt>0):
                T.append(int(t[10:pt]))
            elif(p>=0 and pt<0):
                try:
                    T.append((int(t[11:16])*100)/Price_t[i])
                    print(t[11:16])
                except ValueError as e:
                    try:
                        T.append((int(t[11:15])*100)/Price_t[i])
                        print(t[11:15])
                    except ValueError as e:
                        T.append(5)                                                                                        
            else:
                T.append(0)
        Boff.append(max(T))
        print(T,i,max(T))
            
    else:
        Boff.append(0)
```
